Вам необхідно розрахувати показники за списком щоденних витрат. Деталі реалізації:

* У кодовій комірці створіть список: [1200, 1500, 800, 600, 1000, 400, 900].
* Розрахуйте:
    - загальну суму витрат;
    - середнє значення;
    - максимальна і мінімальна витрата;

```python
spendings = [1200, 1500, 800, 600, 1000, 400, 900]
```

```python
from IPython.display import Markdown
```

```python
Markdown("Index|Value|\n|-|-|\n|Sum of spendings|%.2f|\n|Middle|%.2f|\n|Max spending|%.2f|\n|Min spending|%.2f|\n|Count of days|%.2f|" % (
    sum(spendings),
    sum(spendings) / len(spendings),
    max(spendings),
    min(spendings),
    sum(map(lambda x: 1 if x > 1000 else 0, spendings))
))
```
