# Task 1
Є дані про планові та фактичні продажі за місяцями. Потрібно порівняти два ряди на одному графіку.
* Створіть список місяців і два списки значень: "План" і "Факт".
* Побудуйте один графік із двома лініями.
* Додайте підписи осей і заголовок.
* Додайте легенду для ліній.
* Відобразіть графік на екрані.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('ggplot')
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
plan_sales = [100, 120, 130, 140, 150, 160]
actual_sales = [95, 125, 120, 145, 155, 150]

fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(months, plan_sales, marker='o', linestyle='--', color='blue', label='Plan')
ax.plot(months, actual_sales, marker='s', linestyle='-', color='green', label='Actual')

ax.set_xlabel('Months')
ax.set_ylabel('Sales Volume')
ax.set_title('Monthly Sales: Plan vs Actual')

ax.legend()

fig.tight_layout()
plt.show()
```

# Task 2
Дано значення вікових груп 100 осіб. Потрібно показати розподіл.
* Створіть список (або масив) зі 100 цілих значень віку.
* Побудуйте гістограму розподілу.
* Додайте вертикальну лінію середнього значення.
* Підпишіть осі та додайте заголовок.
* Відобразіть графік на екрані.

```python
ages = np.random.normal(loc=35, scale=10, size=100).astype(int)
ages = np.clip(ages, 18, 80)

fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(ages, bins=10, color='blue', edgecolor='black', alpha=0.7)

mean_age = np.mean(ages)
ax.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label=f'Mean Age ({mean_age:.1f})')

ax.set_xlabel('Age Groups')
ax.set_ylabel('Number of People')
ax.set_title('Age Distribution of 100 Individuals')
ax.legend()

fig.tight_layout()
plt.show()
```

# Task 3
Є результати іспитів за трьома групами студентів. Потрібно порівняти розкид оцінок між групами.
* Створіть три набори числових результатів (по одній вибірці на групу).
* Побудуйте графік boxplot для трьох груп поруч.
* Підпишіть групи по осі X.
* Додайте підписи осей і заголовок.
* Відобразіть графік на екрані.

```python
np.random.seed(42)
group_A = np.random.normal(75, 10, 30)
group_B = np.random.normal(85, 5, 30) 
group_C = np.random.normal(65, 15, 30)

data = [np.clip(group_A, 0, 100), np.clip(group_B, 0, 100), np.clip(group_C, 0, 100)]

fig, ax = plt.subplots(figsize=(8, 5))

ax.boxplot(data, tick_labels=['Group A', 'Group B', 'Group C'], patch_artist=True)

ax.set_xlabel('Student Groups')
ax.set_ylabel('Exam Scores')
ax.set_title('Exam Results Distribution Across Groups')

fig.tight_layout()
plt.show()
```

# Task 5
Є погодинні дані навантаження сервера за добу. Потрібно показати зміни і виділити область між кривою і віссю X.
* Створіть список годин (0-23) і відповідні значення навантаження.
* Побудуйте лінійний графік навантаження.
* Зафарбуйте область під кривою.
* Додайте сітку, підписи осей і заголовок.
* Відобразіть графік на екрані.

```python
hours = np.arange(24)

server_load = [
    22, 18, 15, 12, 14, 20,
    35, 55, 70, 75, 78, 80,
    82, 79, 77, 81, 85, 90,
    95, 92, 85, 60, 45, 30
]

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(hours, server_load, color='blue', linewidth=2, label='CPU Load')

ax.fill_between(hours, server_load, color='blue', alpha=0.4)

ax.grid(True, linestyle='--', alpha=0.6)
ax.set_xticks(hours)
ax.set_xlabel('Hour of Day')
ax.set_ylabel('Server Load (%)')
ax.set_title('Daily Server Load Profile')
ax.legend()

fig.tight_layout()
plt.show()
```

# Task 6
Дано чотири набори метрик продукту: конверсія, утримання, середній чек, кількість замовлень. Потрібно показати їх на окремих панелях.
* Створіть чотири набори числових значень за однією шкалою часу.
* Створіть сітку з чотирьох підграфіків 2×2.
* На кожному підграфіку побудуйте відповідний графік.
* Додайте заголовки для кожного підграфіка та загальний заголовок.
* Відобразіть результат на екрані.

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

conversion_rate = [2.1, 2.3, 2.2, 2.4, 2.8, 3.5, 3.2]
avg_order_value = [120, 115, 118, 125, 140, 165, 150]
total_orders    = [300, 310, 290, 350, 420, 650, 580]

retention_rate  = [45.0, 42.5, 40.1, 38.0, 36.5, 35.0, 34.2]

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].plot(days, conversion_rate, color='green', marker='o')
axs[0, 0].set_title('Conversion Rate %')

axs[0, 1].plot(days, retention_rate, color='orange', linestyle='--', marker='x')
axs[0, 1].set_title('Retention Rate %')

axs[1, 0].plot(days, avg_order_value, color='purple', drawstyle='steps-mid')
axs[1, 0].set_title('Average Order Value')

axs[1, 1].bar(days, total_orders, color='blue', alpha=0.8)
axs[1, 1].set_title('Number of Orders')

fig.suptitle('Weekly Product Metrics Dashboard', fontsize=16)

fig.tight_layout() 
plt.show()
```
