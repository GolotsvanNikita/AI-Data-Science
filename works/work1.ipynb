## Звіт із продажів

У цьому розділі ми рахуємо підсумки за простими даними

```python
sales = [120, 135, 100, 150, 90, 200, 170]
```

```python
from IPython.display import Markdown
```

```python
Markdown("Показник|Значення|\n|-|-|\n|Загальна сума|%.2f|\n|Середнє|%.2f|\n|Максимум|%.2f|\n|Мінімум|%.2f|" % (
    sum(sales),
    sum(sales) / len(sales),
    max(sales),
    min(sales)
))
```
