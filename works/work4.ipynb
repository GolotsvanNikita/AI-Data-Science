```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 11)
```

```python
y = x * x
```

```python
print(x)
print(y)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'o:r')
ax.set_title("Graph of function = x^2")
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
a = 12
b = 2
c = 15

x = np.linspace(-5, 5, 100)

y = a * x**2 + b * x + c

fig, ax = plt.subplots()
ax.plot(x, y, label=f'{a}x^2 + {b}x + {c}')

ax.set_title("Graph of square function")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True)
ax.legend()

plt.show()
```

```python
sales = [150, 200, 250, 300, 350, 280]
months = ['Jan', 'Feb', 'March', 'Apr', 'May', 'Jan']
```

```python
fig, ax = plt.subplots()
ax.bar(months, sales, color='blue')
ax.set_title("Graph of sales since 6 months")
ax.set_xlabel('Month')
ax.set_ylabel('Sale')
plt.show()
```

```python
def to_percent(n): 
    return f'{n:.1f} %'

categories = ['Produсts', 'Clothes', 'Tech', 'Toys']
sales_parts = [40, 30, 55, 20]
fig, ax = plt.subplots()
ax.pie(sales_parts, labels=categories, autopct=to_percent)
plt.show()
```

```python
heights = [160, 165, 170, 175, 180]
weights = [55, 60, 70, 72, 80]

fig, ax = plt.subplots()
ax.grid()
ax.set_title("Addiction weights at heights")
ax.set_xlabel('Heights')
ax.set_ylabel('Weights')
ax.scatter(heights, weights, marker='d', color='green')
plt.show()
```

```python
hours = [0, 4, 8, 12, 16, 20, 24]
temperature = [15, 14, 18, 22, 25, 20, 17]

fig, ax = plt.subplots()
ax.grid()
ax.set_title("Temperature per day")
ax.set_xlabel('Hours')
ax.set_ylabel('Temp')
ax.plot(hours, temperature)
plt.show()
```
