```python
import numpy as np
```

```python
top_limit = int(input('enter the top limit: '))
bottom_limit = int(input('enter the bottom limit: '))
```

```python
matrix = np.random.randint(bottom_limit, top_limit, size=25).reshape(5,5)
print(matrix)
```

```python
main_diag = np.diag(matrix)
print(main_diag)
```

### Sum of the main diagonal elements

```python
print(sum(main_diag))
```

```python
for i in range(1, 5):
    np.fill_diagonal(matrix[:-i, i:], 0)
```
