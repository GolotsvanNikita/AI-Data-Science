* Використовуючи вбудовані засоби Python (наприклад, count(), dict, або collections.Counter), порахуйте, скільки разів зустрічається кожен фрукт.

```python
votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]
```

```python
from IPython.display import Markdown
```

```python
apples = votes.count("apple")
bananas = votes.count("banana")
oranges = votes.count("orange")

fruits = {'apple': apples, 'banana': bananas, 'orange': oranges}

leader = max(fruits.items(), key=lambda x: x[1])[0]
top3 = min(fruits.items(), key=lambda x: x[1])[0]

Markdown(f"Fruits|Count|\n|-|-|\n|{"apple(leader)" if leader == "apple" else "apple"}|%.2f|\n|{"banana(leader)" if leader == "banana" else "banana"}|%.2f|\n|{"orange(leader)" if leader == "orange" else "orange"}|%.2f|" % (
    votes.count("apple"),
    votes.count("banana"),
    votes.count("orange"),
))
```

```python
Markdown(f"# Top 3\n ## {top3}")
```
