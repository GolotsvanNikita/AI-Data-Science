```python
math = [80, 92, 75, 60, 88]
physics = [70, 65, 90, 85, 78]
english = [95, 100, 85, 90, 87]
```

```python
from IPython.display import Markdown
```

```python
mid_math = sum(math) / len(math)
mid_physics = sum(physics) / len(physics)
mid_english = sum(english) / len(english)


Markdown("Subject|Middle grade|Min grade|Max grade|\n|-|-|-|-|\n|Math|%.2f|%.2f|%.2f|\n|Physics|%.2f|%.2f|%.2f|\n|English|%.2f|%.2f|%.2f|" % (
    mid_math,
    min(math),
    max(math),
    mid_physics,
    min(physics),
    max(physics),
    mid_english,
    min(english),
    max(english)
))
```

```python
print(f'Biggest middle grade: {max({'math': mid_math,'physics': mid_physics,'english': mid_english}.items(), key=lambda x: x[1])[0]}')
```
