```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 11)
```

```python
y = x * x
```

```python
print(x)
print(y)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'o:r')
ax.set_title("Graph of function = x^2")
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
a = 12
b = 2
c = 15

x = np.linspace(-5, 5, 100)

y = a * x**2 + b * x + c

fig, ax = plt.subplots()
ax.plot(x, y, label=f'{a}x^2 + {b}x + {c}')

ax.set_title("Graph of square function")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.grid(True)
ax.legend()

plt.show()
```

```python
sales = [150, 200, 250, 300, 350, 280]
months = ['Jan', 'Feb', 'March', 'Apr', 'May', 'Jan']
```

```python
fig, ax = plt.subplots()
ax.bar(months, sales, color='blue')
ax.set_title("Graph of sales since 6 months")
ax.set_xlabel('Month')
ax.set_ylabel('Sale')
plt.show()
```
