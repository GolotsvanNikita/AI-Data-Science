# Task 2
Створіть двовимірний масив розміром 4×4, заповнений випадковими цілими числами від 1 до 50.
* Виведіть масив у читабельному вигляді (у вигляді таблиці).
* Знайдіть суму елементів за кожним рядком і за кожним стовпчиком (використовуйте параметр axis).
* Визначте загальне середнє значення масиву.
* Виведіть усі результати в зручному вигляді
(наприклад: "Суми за рядками: ...","Суми за стовпцями: ...")


```python
rand_4_4 = np.random.randint(1, 50, size=(4,4))
rand_4_4
```

```python
print(np.sum(rand_4_4, axis=0),  # columns
      np.sum(rand_4_4, axis=1),  # rows
      np.mean(rand_4_4))
```

```python
print(*["%s - %d" % (k,v) for (k,v) in zip(rand_4_4, np.sum(rand_4_4, axis=1))], sep='\n')
print(np.sum(rand_4_4, axis=0), "- %f " % np.mean(rand_4_4))
```

# Task 4
* Створіть масив розміром 10×10, заповнений випадковими числами з плаваючою крапкою від 0 до 1.
* Знайдіть мінімальне та максимальне значення в усьому масиві.
* Визначте індекси мінімального та максимального елементів
* (наприклад: (рядок, стовпець)).
* Виведіть сам масив і окремо знайдені значення разом з їхніми індексами.

```python
arr2 = np.random.rand(10, 10)
np.set_printoptions(precision=4)
print(arr2)
# print(f'{arr2[index_max]}, index: {index_max}\n{arr2[index_min]}, index: {index_min}')
```

```python
arr_max = np.max(arr2)
arr_min = np.min(arr2)
print(f'Max: {arr_max}')
print(f'Min: {arr_min}')
arr_max_i = np.argmax(arr2)
arr_min_i = np.argmin(arr2)
print(f'Max index: {arr_max_i}')
print(f'Min index: {arr_min_i}')
max_coord = np.unravel_index(arr_max_i, arr2.shape)
min_coord = np.unravel_index(arr_min_i, arr2.shape)
print(f'Coordinate max element: {', '.join(map(str, max_coord))}')
print(f'Coordinate min element: {', '.join(map(str, min_coord))}')
```

# Task 6
* Створіть два одновимірні масиви довжиною 5: перший заповніть числами від 1 до 5, другий — числами від 6 до 10.
* Об'єднайте їх в один масив за допомогою np.concatenate.
* Побудуйте двовимірний масив 2×5, у якому перший рядок — перший масив, а другий рядок — другий масив.
* Об'єднайте обидва масиви в одну колонку (розмірність 5×2).
* Виведіть усі три результати на екран.


```python
arr5 = np.random.randint(1, 5, size=5)
arr6 = np.random.randint(6, 10, size=5)
conc = np.concatenate((arr5, arr6), axis=0)
print(arr5, arr6)
print(conc)
```

```python
matrix = conc.reshape(2,5)
print(matrix)
```

```python
print(np.transpose(matrix))
```
