```python
from IPython.display import Markdown
```

```python
Markdown("Index|Value|\n|-|-|\n|Middle|%.2f|\n|Maximum|%.2f|\n|Minimum|%.2f|\n|Difference(max-min)|%.2f|" % (
    sum(temperature_list) / len(temperature_list),
    max(temperature_list),
    min(temperature_list),
    max(temperature_list) - min(temperature_list)
))
```
