```python
import numpy as np
```

Запитайте в користувача діапазон значень (нижню і верхню межу), включно з від'ємними числами.
* Створіть одновимірний масив із 15 випадкових цілих чисел у вказаному діапазоні.
* Виведіть вихідний масив.
* Виділіть усі від'ємні елементи та виведіть їх як окремий масив.
* Створіть змінену копію масиву, в якій від'ємні елементи замінено на нулі, і виведіть її.
* Підрахуйте кількість нульових елементів у зміненому масиві та виведіть результат.

```python
low_val = int(input("Enter low: "))
high_val = int(input("Enter high: "))
```

```python
arr4 = np.random.randint(low_val, high_val + 1, 15)
print(arr4)
```

```python
negative_arr = arr4[arr4 < 0]
print('Negative: ', negative_arr)
```

```python
modified_arr4 = np.where(arr4 < 0, 0, arr4)
print('Without negative: ', modified_arr4)
```

```python
zeros_count = np.sum(modified_arr4 == 0)
print('Zeros count: ', zeros_count)
```

# Task 6
Запитайте у користувача розмірність вихідної матриці (наприклад, 4×6).
* Створіть матрицю з послідовно зростаючих цілих чисел заданої розмірності.
* Виведіть вихідну матрицю.
* Запитайте в користувача нові розміри (наприклад, 2×12) і перетворіть матрицю в зазначену форму.
* Виведіть отриману матрицю.
* Визначте мінімальні та максимальні значення в кожному рядку нової матриці та виведіть обидва одновимірні масиви.
* Підрахуйте загальну суму всіх елементів нової матриці та виведіть результат.

```python
r, c = map(int, input("Enter size: ").split('x'))
```

```python
matrix = np.arange(1, r * c + 1).reshape(r, c)
print(matrix)
```

```python
nr, nc = map(int, input("Enter new size: ").split('x'))
reshaped = matrix.reshape(nr, nc)
print(reshaped)
```

```python
print("Min in rows:", reshaped.min(axis=1))
print("Max in rows:", reshaped.max(axis=1))
print("Total sum elements:", reshaped.sum())
```
